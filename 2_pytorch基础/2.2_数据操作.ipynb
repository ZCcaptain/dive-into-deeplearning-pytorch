### 2.2.3 广播机制

对两个形状不同的Tensor做按元素运算可能会触发广播机制：将两个tensor通过复制元素使其形状相同再按元素运算.仅限按元素运算的可以通过复制变成一样的。

```python
x = torch.arange(1, 3).view(1, 2)
print(x)
y = torch.arange(1, 4).view(3, 1)
print(y)
print(x + y)
```

### 2.2.4 运算的内存开销

前面说了，索引、view是不会开辟新内存的，而像y = x + y这样的运算是会新开内存的，然后将y指向新内存。有两种方式可以避免这种内存开销
- 使用y[:]

```python
x = torch.tensor([1, 2])
y = torch.tensor([3, 4])
id_before = id(y)
y[:] = y + x
print(id(y) == id_before) # True
```

- 使用运算全名函数汇总的out参数或者+= 或add_这种inplace操作

```python
x = torch.tensor([1, 2])
y = torch.tensor([3, 4])
id_before = id(y)
torch.add(x, y, out=y) # y += x, y.add_(x)
print(id(y) == id_before) # True
```
